```python
#k-means
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
x_data = np.array([
    [2, 1],
    [4, 4],
    [3, 2],
    [2, 7],
    [1, 8],
    [3, 4],
    [5, 5],
    [7, 5],
    [2, 5],
    [3, 5],
    [8, 9],
    [9, 10],
    [0, 3],
    [6, 12],
    [9, 2],
    [6, 10],
    [2, 4]
])
#임의의 넘파이 어레이 생성
```

```python
import matplotlib.pyplot as plt
```

```python
X=[]
Y=[]
for i in x_data:
  X.append(i[0])
  Y.append(i[1])
```

```python
plt.scatter(X,Y)
#분포 확인
```

```python
model = KMeans(n_clusters=3)
model.fit(x_data)

y_predict = model.predict(x_data)

print(y_predict)
#사이킷런의 kmeans 사용 예제, 클러스터의 갯수를 지정해줘야함.
```

```python
import seaborn as sns

df = pd.DataFrame(x_data)
df['target'] = y_predict
```

```python
sns.pairplot(df, hue='target')
plt.show()
#결과값 분석
```
